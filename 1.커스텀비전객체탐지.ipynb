```python
from azure.cognitiveservices.vision.customvision.prediction import CustomVisionPredictionClient
from msrest.authentication import ApiKeyCredentials
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image

# Custom Vision 설정
PREDICTION_KEY = "PSEXsgMiTa4TsG3QFLsdtEYFGYHDRbiSa8z1ldr6DAwgf8yrp7QEJQQJ99BBACYeBjFXJ3w3AAAIACOGDRt5"  # Azure Portal 확인
ENDPOINT = "https://6a010cv250210-prediction.cognitiveservices.azure.com/customvision/v3.0/Prediction/95bb86f8-bdb7-493d-a20e-67d10ae9e2dd/detect/iterations/smartphone/image"  # Azure Portal 확인
PROJECT_ID = "95bb86f8-bdb7-493d-a20e-67d10ae9e2dd"          # Custom Vision 프로젝트 ID
PUBLISHED_NAME = "smartphone"  # 게시된 Iteration 이름

# Custom Vision 클라이언트 초기화
credentials = ApiKeyCredentials(in_headers={"Prediction-key": PREDICTION_KEY})
predictor = CustomVisionPredictionClient(endpoint=ENDPOINT, credentials=credentials)

# 이미지 예측 함수
def predict_image(image_path):
    with open(image_path, "rb") as image_contents:
        # Custom Vision 호출
        results = predictor.detect_image(PROJECT_ID, PUBLISHED_NAME, image_contents)

    # 결과 출력 및 좌표 (Bounding Box) 반환
    predictions = []
    for prediction in results.predictions:
        if prediction.probability > 0.5:  # 최소 신뢰도 임계값
            bbox = prediction.bounding_box
            predictions.append({
                "tag": prediction.tag_name,
                "probability": prediction.probability,
                "bbox": {
                    "left": bbox.left,
                    "top": bbox.top,
                    "width": bbox.width,
                    "height": bbox.height
                }
            })
    return predictions

# 예측 결과 시각화 함수
def visualize_predictions(image_path, predictions):
    image = Image.open(image_path).convert("RGB")
    fig, ax = plt.subplots(1, figsize=(10, 7))
    ax.imshow(image)

    # 바운딩 박스 시각화
    for prediction in predictions:
        bbox = prediction["bbox"]
        tag = prediction["tag"]
        probability = prediction["probability"]

        # 이미지 크기와 좌표 반영
        left = bbox["left"] * image.width
        top = bbox["top"] * image.height
        width = bbox["width"] * image.width
        height = bbox["height"] * image.height

        # 바운딩 박스 그리기
        rect = patches.Rectangle((left, top), width, height, linewidth=2, edgecolor='r', facecolor='none')
        ax.add_patch(rect)
        ax.text(
            left, top - 10,
            f"{tag} ({probability:.2f})",
            fontsize=12, color='red', backgroundcolor='white'
        )

    plt.axis("off")
    plt.show()

# 사용 예제
image_path = r"C:\Users\mwork\Desktop\prpro\dentdetecting\4.jpg"
predictions = predict_image(image_path)
visualize_predictions(image_path, predictions)
```
